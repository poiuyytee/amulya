```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'x', ms = 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Allensolly", "Burberry", "Calvinklein", "Dior"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "green")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Ants", "Bugs", "Cats", "dogs"]
myexplode = [0, 0, 0.3, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x=range(1,6)
y=[1,4,6,8,4]


plt.fill_between(x, y)
plt.show()
```
